```python
import yfinance as yf

def fetch_financial_data(company='AMZN', period="1y"):
    """
    This function fetches stock market quotations using yfinance.
    The previous attempts with pandas_datareader failed due to RemoteDataError.
    """
    # Download historical market data from Yahoo Finance
    data = yf.download(company, period=period)
    return data

# Fetch data for AMZN for the last year
df = fetch_financial_data(company='AMZN', period="1y")

df.info()
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the 'Close' prices for AMZN
close_prices = df['Close', 'AMZN']

# Create a figure and an axes object
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the close prices
sns.lineplot(x=close_prices.index, y=close_prices.values, ax=ax)

# Add title and labels for clarity
ax.set_title('Ceny zamknięcia AMZN w czasie')
ax.set_xlabel('Data')
ax.set_ylabel('Cena zamknięcia')
ax.grid(True, linestyle='--', alpha=0.7)

# Rotate x-axis labels for better readability if dates overlap
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```
